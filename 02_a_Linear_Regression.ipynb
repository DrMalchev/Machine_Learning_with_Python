```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0,10.0)

#Reading data
data = pd.read_csv('headbrain.csv')
print(data.shape)
data.head()
```

```python
# Collect x and Y
x = data['Head Size(cm^3)'].values
y = data['Brain Weight(grams)'].values
```

```python
# Calc mean value

mean_x = np.mean(x)
mean_y = np.mean(y)

# Total number of values
n = len(x)

# Calc b0 and b1
numer = 0
denom = 0

for i in range(n):
    numer+=(x[i] - mean_x)*(y[i] - mean_y)
    denom+=(x[i] - mean_x)**2
    b1 = numer/denom
    b0 = mean_y - (b1*mean_x)

print(b1,b0)
```

```python
# Plotting

max_x = np.max(x)+100
min_x = np.min(x)-100

# calc line values
xaxis = np.linspace(min_x, max_x, 1000)
yaxis = b0 + b1*xaxis

# Plot regression line
plt.plot(xaxis, yaxis, color='r', label = 'Regresion Line')
# Plot raw data
plt.scatter(x,y,color='b', label = 'Data')

plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.legend()
plt.show()
```

```python
# Quality check using least square method

ss_t=0
ss_r=0
for i in range(n):
    y_pred = b0 + b1*x[i]
    ss_t+=(y[i] - mean_y)**2
    ss_r+=(y[i] - y_pred)**2
r2=1-ss_r/ss_t
print(r2)
```
