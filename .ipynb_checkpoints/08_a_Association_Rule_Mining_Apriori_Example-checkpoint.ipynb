```python
# In[*] means the kernel is busy
# in this case just wait for it to execute
# It takes several minutes -> WAIT!
#
df=pd.read_excel("Online_Retail.xlsx",sheet_name="Online_Retail")
df.head()
```

```python
#Remove spaces from beginning and end of string
df['Description']= df['Description'].str.strip()
#Remove data without InvoiceNo
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo']= df['InvoiceNo'].astype('str')
df=df[~df['InvoiceNo'].str.contains('c')]
df
```

```python
# Use only data from France
basket = (df[df['Country']=='France']
         .groupby(['InvoiceNo','Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo')
         )
basket
```
